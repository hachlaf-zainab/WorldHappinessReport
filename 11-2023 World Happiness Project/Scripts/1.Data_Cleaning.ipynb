## 03. Data consistency checks

```python
# Check for mixed-type data in 'Latest_ranking'

for col in Latest_ranking.columns.tolist():
  weird = (Latest_ranking[[col]].applymap(type) != Latest_ranking[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (Latest_ranking[weird]) > 0:
    print (col)
```

```python
# Check for mixed-type data in 'All_rankings'

for col in All_rankings.columns.tolist():
  weird = (All_rankings[[col]].applymap(type) != All_rankings[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (All_rankings[weird]) > 0:
    print (col)
```

### There are none in both data frames.
